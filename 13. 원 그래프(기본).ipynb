# 13. 원 그래프 (기본)

```python
import matplotlib.pyplot as plt 
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows 버전 글꼴 변경
matplotlib.rcParams['font.size'] = 15 #글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False #한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
values = [30,25,20,13,10,2]
labels = ['Python' , 'Java' , 'Javascript' ,'C#','C/C++','ETC']

plt.pie(values , labels=labels, autopct='%.1f%%',startangle=90)
plt.show()
```

```python
values = [30,25,20,13,10,2]
labels = ['Python' , 'Java' , 'Javascript' ,'C#','C/C++','ETC']
explode=[0.2,0.1,0,0,0,0] #간격 벌리기

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
values = [30,25,20,13,10,2]
labels = ['Python' , 'Java' , 'Javascript' ,'C#','C/C++','ETC']
explode=[0.05]*6  #0.05의 간격 만큼 6개의 리스트가 만들어진다.

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.title('언어별 선호도')
plt.legend(loc=(1.2,0.3), title='언어') #범례 표시 1.2,0.3 만큼 옮긴다 loc=위치, 튜플 형태로
plt.show()
```
